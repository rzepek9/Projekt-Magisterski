```python
colors = {
    0: 'red',
    1: 'black',
    2: 'green'
}
parts = ['SrodekBiodra', 'RHip', 'RKnee', 'RAnkle', 'LHip', 'LKnee', 'LAnkle', 'Stomach', 'Neck', 'Head', 'Nose', 
         'LShoulder', 'LElbow', 'LWrist', 'RShoulder', 'RElbow', 'RWrist']
```

```python
def plot_signals(signal):
    fig, axs = plt.subplots(9, 2, figsize=(18, 18))
    fig.tight_layout(pad=4.0)
    axs = axs.flatten()
    z = 0
    signal = signal.transpose(1, 0)
    part_id = 0
    for i, channel in enumerate(signal):
        i += 1
        if i % 3 != 0:
            timestep = [i for i in range(channel.shape[0])]
            axs[part_id].plot(timestep, channel, color=colors[i%3], linewidth = 3)
        
        else:
            axs[part_id].plot(timestep, channel, color=colors[i%3], linewidth = 3)
            axs[part_id].set_xlabel('timestep')
            axs[part_id].set_ylabel('values')
            axs[part_id].set_title(parts[part_id])
            axs[part_id].imshow
            
            part_id += 1
        
```

```python
signal, _, file_name = dataset[0]
```

```python
signal_np = signal.numpy()
```

```python
plot_signals(signal_np)
```
